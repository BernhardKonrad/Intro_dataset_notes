# Data set suggestion: Musical notes for image recognition

Image recognition is a classical and challenging task in machine learning. The data set at hand consists of small images of musical notes. The task is to classify each image of a note to its corresponding pitch.

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(7,4))
fig.tight_layout()

ax1.imshow(Image.open('e.jpg'), cmap='gray')
ax1.set_title('Note e')

ax2.imshow(Image.open('g.jpg'), cmap='gray')
ax2.set_title('Note g')

ax3.imshow(Image.open('a.jpg'), cmap='gray')
ax3.set_title('Note a')
                                         
ax4.imshow(Image.open('c.jpg'), cmap='gray')
ax4.set_title('Note c')
```

The images come in `jpg` format of size `80 * 32 = 2560` pixels, where each pixel represents a gray value between `0` and `255`.

Each image can be represented by a matrix, and we can even recognize the note if we zoom in correctly.

```python
note = np.asarray(Image.open('g.jpg'))
print('Shape of each image:', note.shape)
print(note[35:55,10:22])
```

Notes can be of [various durations](https://en.wikipedia.org/wiki/List_of_musical_symbols#Notes_and_rests), and may be beamed or dotted. [Accidentals](https://en.wikipedia.org/wiki/List_of_musical_symbols#Accidentals_and_key_signatures) can be ignored. The full dataset is about `6400` images that belong to a total of `22` (unbalanced) classes.

Ultimately, a good classifier might enable a computer to understand hand-written musical sheets.
As a starter project, students can apply simple machine learning algorithms like logistic regression to separate the two largest classes. This classifier can then be extended to all classes (e.g. one-vs-all). Ultimately, though, a [convolutional neural net](https://en.wikipedia.org/wiki/Convolutional_neural_network) should be best suited for this kind of image recognition task.

A stretch goal would be to not only recognize the pitch, but also understand the duration.
